## <font color = lightblue>Dicionário de Variáveis
Estrutura do dicionário 'dados'

`dados`: Um dicionário contendo informações sobre as atividades realizadas em cada dia da semana.
* `Dias`: Lista com os nomes dos dias da semana (segunda a domingo).
* `Horas` Trabalhadas': Lista com o número de horas trabalhadas em cada dia.
* `Bugs Corrigidos`: Lista com a quantidade de bugs corrigidos em cada dia.
* `Tarefas Concluídas`: Lista com o número de tarefas concluídas em cada dia.

`DataFrame`: Um DataFrame criado usando a biblioteca pandas a partir do dicionário "dados", contendo os seguintes campos:
* `Dias`: Dias da semana.
* `Horas Trabalhadas`: Total de horas trabalhadas em cada dia.
* `Bugs Corrigidos`: Total de bugs corrigidos em cada dia.
* `Tarefas Concluídas`: Total de tarefas concluídas em cada dia.
* `Produtividade Diária`: Média diária de tarefas concluídas por hora trabalhada em cada dia.

`total_horas_trabalhadas`: Armazena o total de horas trabalhadas em toda a semana.

`media_diaria_horas_trabalhadas`: Armazena a média diária de horas trabalhadas com duas casas decimais.

`total_bugs_corrigidos`: Armazena o total de bugs corrigidos em toda a semana.

`media_diaria_bugs_corrigidos`: Armazena a média diária de bugs corrigidos com duas casas decimais.

`total_tarefas_concluidas`: Armazena o total de tarefas concluídas em toda a semana.

`media_diaria_tarefas_concluidas`: Armazena a média diária de tarefas concluídas com duas casas decimais.

`produtividade_diaria`: Uma série que representa a produtividade diária (tarefas concluídas por hora trabalhada) para cada dia.

```python
# Nesta parte do código, é importado as bibliotecas pandas e matplotlib.
import pandas as pd
import matplotlib.pyplot as plt
```

<br>

---
-

```python
# Nesta parte do código, é definindo um dicionário chamado dados
# que contém informações sobre as atividades realizadas em cada dia da semana.
dados = {
    'Dias': ['segunda', 'terça', 'quarta', 'quinta', 'sexta', 'sábado', 'domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 6],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

# Nesta parte do código, é utilizado a biblioteca pandas para criar um DataFrame a partir do dicionário dados.
DataFrame = pd.DataFrame(dados)
```

<br>

---
-

<br>

---
-

```python
# Nesta parte do código, estamos calculando o total de horas trabalhadas dos os dias da semana.
total_horas_trabalhadas = DataFrame['Horas Trabalhadas'].sum()
print(f'\n\033[1;4;36mTotal de Horas Trabalhadas: {total_horas_trabalhadas}\n')

# Nesta parte do código, é calculado a média diária de horas trabalhadas e imprimindo o resultado formatado com duas casas decimais.
media_diaria_horas_trabalhadas = DataFrame['Horas Trabalhadas'].mean()
print(f'Média Diaria de horas Trabalhadas: {media_diaria_horas_trabalhadas:.2f}\n')

# Gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(DataFrame['Dias'], DataFrame['Horas Trabalhadas'])
plt.title('Horas Trabalhadas por Dia da Semana')
plt.xlabel('Dias')
plt.ylabel('Total de Horas Trabalhadas')
plt.grid(True)
plt.show()
```

```python
# Nesta parte do código, é calculado o total de bugs corrigidos dos dias da semana.
total_bugs_corrigidos = DataFrame['Bugs Corrigidos'].sum()
print(f'\n\033[1;4;36mTotal de Bugs Corrigidos por Semana: {total_bugs_corrigidos}\n')

# Nesta parte do código, é calculado a média diária de bugs corrigidos e imprimindo o resultado formatado com duas casas decimais.
media_diaria_bugs_corrigidos = DataFrame['Bugs Corrigidos'].mean()
print(f'Média Diaria de Bugs Corrigidos por Semana: {media_diaria_bugs_corrigidos:.2f}\n')

# Gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(DataFrame['Dias'], DataFrame['Bugs Corrigidos'])
plt.title('Bugs Corrigidos por Dia da Semana')
plt.xlabel('Dias')
plt.ylabel('Total de Bugs Corrigidos')
plt.grid(True)
plt.show()
```

```python
# Nesta parte do código, é calculando o total de tarefas concluídas em todos os dias da semana.
total_tarefas_concluidas = DataFrame['Tarefas Concluídas'].sum()
print(f'\n\033[1;4;36mTotal de Tarefas Comcluídas: {total_tarefas_concluidas}\n')

# Nesta parte do código, é calculado a média diária de tarefas concluídas imprimindo o resultado formatado com duas casas decimais.
media_diaria_tarefas_concluidas = DataFrame['Tarefas Concluídas'].mean()
print(f'Média Diaria de Tarefas Comcluídas: {media_diaria_tarefas_concluidas:.2f}\n')


# Gráfico de barras para o total de tarefas concluídas
plt.figure(figsize=(8, 6))
plt.bar(DataFrame['Dias'], DataFrame['Tarefas Concluídas'])
plt.title('Total de Tarefas Concluídas por Dia da Semana')
plt.xlabel('Dias')
plt.ylabel('Total de Tarefas Concluídas')
plt.grid(True)
plt.show()
```

## Analises que poderm ser realizadas
- Variação da Produtividade ao Longo da Semana
- Identificação de Dias Mais Produtivos
- Dias com Baixa Produtividade
- Comparação com Outras Métricas
- Impacto das Horas Trabalhadas
- Feedback e sugestões de melhorias Melhorias



```python
# Nessa parte do código, divide o número de "Tarefas Concluídas" pelo número de "Horas Trabalhadas" para cada dia da semana no DataFrame.
produtividade_diaria = DataFrame['Tarefas Concluídas'] / DataFrame['Horas Trabalhadas']
print(f'\n\033[1;36mProdutividade Diaria:\n{produtividade_diaria}')

# Adicionar a coluna "Produtividade Diária" ao DataFrame
DataFrame['Produtividade Diária'] = produtividade_diaria

# Gráfico de Produtividade Diária (Tarefas Concluídas por Hora)
plt.figure(figsize=(8, 6))
plt.plot(range(len(DataFrame)), DataFrame['Produtividade Diária'], marker='o', linestyle='-')
plt.xticks(range(len(DataFrame)), DataFrame['Dias'])
plt.title('Produtividade Diária (Tarefas Concluídas por Hora)')
plt.xlabel('Dias')
plt.ylabel('Produtividade Diária (Tarefas/Hora)')
plt.grid(True)
plt.show()
```

Variação da Produtividade ao Longo da Semana: Utilizando o gráfico de linha que mostra a produtividade diária (tarefas concluídas por hora) para cada dia da semana, observe se há alguma tendência de aumento ou diminuição da produtividade em determinados dias. Isso pode ajudar a identificar quais dias da semana a equipe é mais ou menos produtiva.

Identificação de Dias Mais Produtivos: Verifique quais dias da semana apresentam a maior produtividade diária média. Isso pode ajudar a identificar os dias em que a equipe é mais eficiente e planejar atividades importantes para esses dias.

Dias com Baixa Produtividade: Identifique os dias da semana com a menor produtividade diária média. Isso pode ser um indicativo de possíveis gargalos ou problemas no processo de desenvolvimento que precisam ser abordados.

Comparação com Outras Métricas: Compare a produtividade diária com outras métricas, como horas trabalhadas, bugs corrigidos ou tarefas concluídas. Isso pode ajudar a entender se há alguma relação entre essas métricas e se a produtividade está relacionada ao esforço investido ou a outras variáveis.

Impacto das Horas Trabalhadas: Analise como a produtividade diária varia em relação às horas trabalhadas em cada dia. Isso pode indicar se a equipe é mais produtiva em dias com maior ou menor carga horária.

Análise Qualitativa: Além das métricas quantitativas, considere realizar entrevistas ou pesquisas com a equipe para obter insights qualitativos sobre os fatores que influenciam a produtividade diária. Aspectos como motivação, ambiente de trabalho e suporte técnico podem desempenhar um papel significativo na produtividade.

Feedback e Melhorias: Utilize os resultados da análise de produtividade diária para fornecer feedback à equipe e identificar oportunidades de melhorias. Com base nas descobertas, ajuste as estratégias de trabalho ou implemente mudanças para aumentar a produtividade ao longo da semana.
